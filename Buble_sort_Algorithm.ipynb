```python
jersey=football['Jersey Number']
```

Now sort that column in ascending order

```python
for i in range(len(jersey)-1,0,-1):
    for j in range(i):
        if(jersey[j]>jersey[j+1]):
            temp=jersey[j]
            jersey[j]=jersey[j+1]
            jersey[j+1]=temp
#Display the column
print(jersey.head(5))
```
